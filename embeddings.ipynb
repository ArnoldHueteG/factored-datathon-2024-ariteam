## Generate Embeddings
Generar embeddings

```python
# Función para generar embeddings
from openai import OpenAI
client = OpenAI()

def get_embedding(text, model="text-embedding-3-small"):
   text = text.replace("\n", " ")
   print(text)
   try: 
      return client.embeddings.create(input = [text], model=model).data[0].embedding
   except:
      return 0


# Aplicar la función para generar y agregar la columna de embeddings
data_news['embedding'] = data_news['summary'].apply(lambda x: get_embedding(x))

# Mostrar el DataFrame con la nueva columna
print(data_news)
```

```python
# Run chain

from langchain.chat_models import ChatOpenAI

llm = ChatOpenAI(temperature=0, model="gpt-4o-mini")
chain = create_extraction_chain(event_type_schema, llm)
```

```python
extracted_data = data_news['summary'].apply(lambda x: chain.run(x)[0]).apply(pd.Series)
```
